```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import torch
from torch.cuda import is_available
```

```python
x = np.arange(0, 50, 1)
y = x**(1/3)
y2 = -0.7*x + 2
```

```python
plt.figure(figsize=(10, 4))
plt.plot(x, y, color='red', label='red')
plt.plot(x, y2, color='blue', label='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sample graph')
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```
